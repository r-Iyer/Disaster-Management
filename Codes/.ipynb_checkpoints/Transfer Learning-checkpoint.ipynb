```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
earthquake=load_images_from_folder("../Dataset/Earthquake")
flood=load_images_from_folder("../Dataset/Flood")
other=load_images_from_folder("../Dataset/Other")
```

```python
earthquake=np.array(earthquake)
flood=np.array(flood)
other=np.array(other)
```

```python
data=np.concatenate((earthquake,flood,other),axis=0)
len(data)
```

```python
y=np.zeros(len(data))
y[len(earthquake):(len(earthquake)+len(flood))]=1
y[(len(earthquake)+len(flood)):(len(earthquake)+len(flood)+len(other))]=2
len(y)
```

```python
data = data.astype('float32')
data/=255
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
y= np_utils.to_categorical(y)
```

# Splitting to train,test with balanced classes

```python
from sklearn.model_selection import StratifiedShuffleSplit
stratSplit = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
for train_idx, test_idx in stratSplit.split(data, y):
    X_train=data[train_idx]
    Y_train=y[train_idx]
    X_test=data[test_idx]
    Y_test=y[test_idx]
```

```python
def sliding_window(image, stepSize, windowSize):
	# slide a window across the image
	for y in range(0, image.shape[0], stepSize):
		for x in range(0, image.shape[1], stepSize):
#			print(y + windowSize[1])
			yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
```

```python
steps=48
img1= np.pad(img1, pad_width=steps, mode='constant', constant_values=0)
img1=img1[:,:,steps:steps+3]
mask=np.zeros(np.asarray(img1).shape[:2])
mask.shape
```
